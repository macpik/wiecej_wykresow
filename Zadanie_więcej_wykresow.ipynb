```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]
```

```python
df = pd.DataFrame(prices)
df
```

```python
df = pd.DataFrame(prices, columns = ["month", "price_PLN"])
df = df.set_index("month")
```

```python
df
```

```python
df['price_USD'] = df['price_PLN'].apply(lambda price_PLN: price_PLN / 4)
```

```python
df
```

```python
df["price_USD"].plot(title = "price in usd", style='--r')
```
